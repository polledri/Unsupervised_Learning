# loi normale centrée réduite multivariée (dimension 2)

```python
vg_2d=VG([0,0],[[1,0],[0,1]])
vg_2d.simulation(10000)
vg_2d.plot_grah()
```

Le graphique du dessus montre un tirage de 10000 réalisation d'une loi normale multivariée centrée réduite. La moyenne théorique est donc le vecteur [0,0]. 

## Calcul de la moyenne empirique

On souhaite calculer de facon empirique la moyenne de notre loi multivarié en utilisant la méthode de Monte-Carlo.

```python
vg_2d.compute_empirical_mean(first_n=1000)
```

La moyenne empirique est bien un estimateur de la moyenne.

# Loi normale de moyenne choisi au hasard et de matrice de covariance arbitraire

On décide de refaire un exmple avec une moyenne [5,2] et une matrice de covariance [[0.1,3],[3,-2]]

```python
vg_2d=VG([5,2],[[0.1,-3],[-3,10]])
vg_2d.simulation(10000)
vg_2d.plot_grah()
```

Ici on n'observe un jeu de donnée répartit dans une ellipse centrée en [5,2]. La forme de l'ellipse est dû à la matrice de covariance pour les positions 1,2 et 2,1 : une forte covariance entre les lois implique un affinage du cercle. le choix des cotés affinées va ensuite dépendre du signe des valeurs en position 1,2 et 2,1. 

```python
vg_2d.compute_empirical_mean(first_n=1000)
```

On remarque de même une convergence de la moyenne empirique, mais cette fois-ci le temps de convergence prend plus de temps à cause de la variance des lois X et Y. (rf intervalle de confiance de loi gaussienne multivarié)
