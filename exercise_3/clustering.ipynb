```python
from algo_clust import Clustering_supervise
import numpy as np
```

```python
data = np.load('data.npy')
```

# Visualisation des Données sur des sous espaces projetés

```python
clust_sup=Clustering_supervise(data)
```

```python
clust_sup.plot_distribution(0,2)
```

On remarque alors que le clustering va etre très simple (un simple k plus proche voisin suffira en utilisant cette dimension). Mais pour faire complexe on choisit d'abord d'utiliser dbscan car on ne choisit pas le nombre de cluster.

# Clustering avec DBscan

```python
clust_sup.dbscan_fit(eps=0.3,min_samples=10)
clust_sup.compute_calinski_harabasz_score(db_or_optics="dbscan")
clust_sup.compute_silhouette_score(db_or_optics="dbscan")
```

# Clustering avec Optics

```python
del clust_sup
clust_sup=Clustering_supervise(data)
clust_sup.optics_fit(min_samples=50, xi=0.05, min_cluster_size=0.05)
clust_sup.compute_calinski_harabasz_score(db_or_optics="optics")
clust_sup.compute_silhouette_score(db_or_optics="optics")
```

Les deux algorithmes sont très puissants et réussissent à classifier le jeux de données avec facilité, c'est ce que montre les deux indicateurs : le score de Calinski Harbasz et le score silhouette. Malheureusement on ne peut pas se rendre compte de leur puissance car le jeu de donnée est trop simple. Un simple kmeans aurait suffit
