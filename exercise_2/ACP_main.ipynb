```python
from CustomPCA import CustomPCA
import numpy as np

data = np.load('data.npy')
labels = np.load('labels.npy')
```

# ACP en dimension 2

On effectue une ACP de dimension 2 sur les donn√©es

```python
pca_2d = CustomPCA(2)
pca_2d.simulation(data, labels)
```

La projection des donn√©es sur l'espace engendr√© par les vecteurs propres dont les valeurs propres sont les deux plus grandes du spectre de la matrice de covariance associ√© au jeu de donn√©e n'est pas satisfaissante car on observe un chevauchement des deux classes, ce qui rend la classification dans cette espace non consistante.

Le graphique des valeurs propres rang√©s par ordre d√©croissante nous montre la d√©croissance lin√©aire. il est donc tr√®s conseill√© de faire une ACP en dimension 3 pour permettre une meilleure classification.

```python
print(pca_2d.pca.explained_variance_ratio_)
```

```python
pca_2d.plot_explained_variance()
```

# ACP en dimension 3

Comme la part des 2 premi√®re valeurs propres dans le spectre de la matrice de covariance ne repr√©sente que ... on peut regarder ce que donne la m√™me m√©thode en rajoutant une dimension aux sous espaces de dimension 2 pr√©c√©dant dont le vecteur propre associ√© √† la 3 √®me plus grande valeur propre est le vecteur directeur.

```python
pca_2d = CustomPCA(3)
pca_2d.simulation(data, labels)
```

On remarque alors que l'apport des 3 premi√®re valeur propre est tr√®s satisfaisant. De plus les donn√©es projet√©es sur les sous espace de dimension 3 permet une classification par hyperplan !
